```python
import pandas as pd
df = pd.read_csv('/home/nayan/Desktop/ACI_MLE/Data/ROI_Corr_FFPS.csv')
```

```python
df
```

```python
corr_mat = df.corr(method='pearson')
corr_mat['ROI']
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming you have a DataFrame called 'data' with your dataset
correlation_matrix = corr_mat["ROI"]

# Set the figure size
plt.figure(figsize=(10, 8))

# Create the heatmap
sns.heatmap(correlation_matrix.to_frame(), annot=True, cmap='RdYlBu', vmin=-1, vmax=1)

# Add a title
plt.title('Correlation Matrix')

# Show the plot
plt.show()
```
